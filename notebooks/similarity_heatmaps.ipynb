```python
# let's try normalizing the edit distances

# https://stackoverflow.com/questions/45783385/normalizing-the-edit-distance
# http://www.cse.lehigh.edu/~lopresti/Publications/1996/sdair96.pdf
def normalized_edit_similarity(m, d):
    # d : edit distance between the two strings
    # m : length of the shorter string
    return ( 1.0 / math.exp( d / (m - d) ) )

# TO DO: don't currently have enough time to get the length of the shorter string, so not normalizing for now
```
